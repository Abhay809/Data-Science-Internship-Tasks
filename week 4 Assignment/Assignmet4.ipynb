```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("penguins.csv")
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].hist(
    bins=20, edgecolor='black', figsize=(10, 6))
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='species', y='body_mass_g', palette="Set2")
plt.title("Body Mass by Species")
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='species', y='bill_length_mm', palette="Set3")
plt.title("Bill Length by Species")
```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
corr = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numerical Features")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='bill_length_mm', y='body_mass_g', hue='species', style='sex')
plt.title("Bill Length vs Body Mass")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Body Mass (g)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 4))
```

counting plots

```python
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='species', palette='pastel')
plt.title("Species Distribution")
```

```python
plt.subplot(1, 3, 2)
sns.countplot(data=df, x='island', palette='muted')
plt.title("Island Distribution")
```
